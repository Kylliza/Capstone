```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pokemon_weaknesses = pd.read_csv('7_types.csv', index_col=0)

# convert the values to numeric
pokemon_weaknesses = pokemon_weaknesses.apply(pd.to_numeric, errors='coerce')

# transpose the dataframe
pokemon_weaknesses = pokemon_weaknesses.transpose()

# plot a horizontal bar chart
ax = pokemon_weaknesses.plot(kind='barh', stacked=True, color=['red', 'green'], figsize=(10, 10))

# set the title and labels
ax.set_title('Pokemon Weaknesses and Resistances')
ax.set_xlabel('Count')
ax.set_ylabel('Type')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load the dataset
pokemon_weaknesses = pd.read_csv('7_types.csv', index_col=0)

# convert the values to numeric
pokemon_weaknesses = pokemon_weaknesses.apply(pd.to_numeric, errors='coerce')

# create a correlation matrix
corr_matrix = pokemon_weaknesses.corr()

# create the heatmap
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt='.1f', center=0)

# add labels and title
plt.title('Pokemon Weaknesses and Resistances Correlation')
plt.xlabel('Pokemon Types')
plt.ylabel('Pokemon Types')

# show the plot
plt.show()

# need to change the axis
```
